```python
def get_data_str(text):
    return [elem[2:].split(',') for elem in text.split()]
```

```python
get_data_str(data_test)
```

```python
def get_quadrant(x, y, vx, vy, steps, size_x = 101, size_y = 103):
    pos_x, pos_y = (x+steps*vx)%size_x, (y+steps*vy)%size_y
    if pos_x == size_x//2 or pos_y == size_y//2:
        return 0
    return 1+2*int(pos_x/(1+size_x//2))+ int(pos_y/(1+size_y//2))
```

```python
def main():
    result = {1:0, 2:0, 3:0, 4:0}
    for elem in data.split('\n')[:-1]:
        line = get_data_str(elem)
        quad = get_quadrant(int(line[0][0]), int(line[0][1]), int(line[1][0]), int(line[1][1]), 100)
        if quad != 0:
            result[quad] += 1
    res = 1
    for val in result.values():
        res *= val
    return res
```

```python
main()
```

```python
def get_pos(x, y, vx, vy, steps, size_x = 101, size_y = 103):
    return (x+steps*vx)%size_x, (y+steps*vy)%size_y
```
